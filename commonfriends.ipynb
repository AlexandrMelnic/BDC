```python
import pyspark
```

```python
sc = pyspark.SparkContext(appName='CommonFriends')
```

```python
file = sc.textFile('soc-LiveJournal1Adj.txt')
```

```python
def separate_edges(adj):
    root = adj[0]
    result = []
    for elem in adj[1].split(','):
        result.append((elem, [root]))
    return result

def separate_list(elem):
    result = []
    for i in range(len(elem[1])-1):
        for j in range(i+1,len(elem[1])):
            result.append(((elem[1][i], elem[1][j]),1))
    return result
```

```python
adj_list = file.map(lambda x: x.split()).filter(lambda x: len(x)>1).flatMap(separate_edges)
adj_list = adj_list.reduceByKey(lambda x,y : x + y ).flatMap(separate_list).reduceByKey(lambda x,y : x+y)
```

```python
adj_list.collect()
```
